```python
import glob
import yaml
import torch
import numpy as np
import torch.nn as nn
import PIL.Image as pil
import matplotlib.pyplot as plt

from torchvision.transforms import ToTensor
```

```python
with open("config.yaml", "r") as file:
    cfg = yaml.safe_load(file)

cfg
```

```python
data_root = cfg["data_root"]
images = sorted(glob.glob(data_root + "*.png"))
images
```

```python
imgs = []
for im in images:
    if 'grayscale' in im:
        continue
    image = pil.open(im)
    image = image.convert('RGB')
    image = np.array(image, dtype=np.float32)
    image = ToTensor()(image).unsqueeze(0)
    if torch.backends.mps.is_available():
        print("MPS available")
        image.to(torch.device("mps"))
    print(image.shape)
    imgs.append(image)
```

# Experiment 1: 2D Convolutions
| S/N | In Channels | Out Channels | kernel size |
| --- | --- | --- | --- |
| 1 | 3 | 3 | 1 |

```python
in_chs = cfg["in_chs"]
out_chs = cfg["out_chs"]
kernels = cfg["kernels"]

print(kernels)
```

```python
convs = []
for kernel_size in kernels:
    conv2d = nn.Conv2d(in_chs, out_chs, kernel_size)
    if torch.backends.mps.is_available():
        print("MPS available")
        conv2d.to(torch.device("mps"))
    nn.init.constant_(conv2d.weight.data, 0.1)
    nn.init.constant_(conv2d.bias.data, 0.0)
    convs.append(conv2d)
```

```python
for input in imgs:
    for idx, conv2d in enumerate(convs):
        try:
            output = conv2d(input)
        except NotImplementedError:
            conv2d.to(torch.device("cpu"))
            output = conv2d(input)
        
        output = output.squeeze()
        output = output.permute((1, 2, 0))
        output = output.detach().numpy().squeeze()
    
        normalized_output = (output - output.min()) / (output.max() - output.min())
    
        plt.figure(figsize=(20, 30))
        plt.imshow(normalized_output)
        pil_image = pil.fromarray(np.uint8(normalized_output * 255))
        pil_image.save(f"outputs/kernel_image_{idx}.jpg")
        
        pil_image2 = pil.fromarray(np.uint8(output))
        pil_image2.save(f"outputs/kernel_features_{idx}.jpg")
```
